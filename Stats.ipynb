```python
import glob
import pandas as pd
x = []
for i in glob.glob("*/*.csv"):
    print (i)
    df = pd.read_csv(i)
    ids = [j.split('&')[0].strip('!') for j in df[df.ID.notnull()]['ID']]
    for k in ids:
        x.append(k)
    #break
```

```python
len(x)
```

```python
from py_youtube import Data
views=0
for ii,i in enumerate(x):
    try:
        link = 'https://www.youtube.com/watch?v='+str(i)
        view = int(Data(link).data()['views'])
        views+=view
        print(ii,i,link,view)
    except:
        print ('Failed for', ii,i,link )
        pass
```

```python
import glob

x = []
words = []
for i in glob.glob("*/Video*.md"):
    x.append(i)
    f = open(i, "r")
    lines = f.read().splitlines()
    f.close()
    for j in lines:
        for k in j.split():
            words.append(k)
print(words[0:10])
print("Number of videos:", len(x))
print("Number of words:", len(words))
```
